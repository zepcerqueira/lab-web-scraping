#### 2.2. Display the top 10 languages by number of native speakers stored in a pandas dataframe.
Hint: After finding the correct table you want to analyse, you can use a nested **for** loop to find the elements row by row (check out the 'td' and 'tr' tags). <br>An easier way to do it is using pd.read_html(), check out documentation [here](https://pandas.pydata.org/pandas-docs/version/0.23.4/generated/pandas.read_html.html).

````python
# This is the url you will scrape in this exercise
url4 = 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'
````

````python
# your code here

status4 = pd.read_html(url4)

language_table = status4[0]

language_table.head(10)
````

````python

status4 = requests.get(url4)

wiki_lang_top_soup = BeautifulSoup(status4.content)

top_lang_wiki = wiki_lang_top_soup.find("table", class_="wikitable")

wiki_languages_data = []

wiki_rows = top_lang_wiki.find_all('tr')[1:] 

for row in wiki_rows:
    columns = row.find_all('td')
    
    language = columns[0].find('a').get_text()
    native_speakers = str(columns[1].get_text().strip())
    
    wiki_languages_data.append({'Language': language, 'Native Speakers (millions)': native_speakers})

wiki_languages_data.sort(key=lambda x: x['Native Speakers (millions)'], reverse=True)

top_10_wiki_lang = wiki_languages_data[:10]
````

#### 3. Display IMDB's top 250 data (movie name, initial release, director name and stars) as a pandas dataframe.
Hint: If you hover over the title of the movie, you should see the director's name. Can you find where it's stored in the html?

````python
# This is the url you will scrape in this exercise 
url5 = 'https://www.imdb.com/chart/top'
````

````python
import re
````

````python
# your code here

movie_names = []
release_data = []
years = []
star = []
ratings = []

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36'
}

status5 = requests.get(url5, headers=headers)
movies_soup = BeautifulSoup(status5.content)
movie_table_list = movies_soup.find_all("ul", class_="ipc-metadata-list")

for component in movie_table_list:
    
    cells = component.find_all("td")
    #movies Names
    movie_name = component.find_all("h3", class_="ipc-title__text")  
    
    #movies release dates
    release = component.find_all("div", class_="sc-4dcdad14-7")  
    
    #movies stars
    description_box = component.find_all("span", class_="ipc-rating-star")
    
    for name in movie_name:
        movie_names.append(name.text.strip())
        
    for date in release:
        release_data.append(date.text.strip())

    for discrip in description_box:
        star.append(discrip.text.strip())


for date in release_data:
    match = re.search(r'\d{4}', date)
    if match:
        year = match.group()
        years.append(year)
        

for item in star:
    match = re.search(r'\d+\.\d+', item)
    if match:
        rating = match.group()
        ratings.append(rating)
        
        
````

#### 3.1. Display the movie name, year and a brief summary of the top 10 random movies (IMDB) as a pandas dataframe.

````python
#This is the url you will scrape in this exercise
url6 = 'https://www.imdb.com/list/ls009796553/'
````

````python
# your code here
movie_names = []
movie_years = []
summary = []


headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36'
}

status5 = requests.get(url6, headers=headers)
movies_soup = BeautifulSoup(status5.content)
movie_table_list = movies_soup.find_all("div", class_="lister-list")

for component in movie_table_list:
    movie_data = component.find_all("h3", class_="lister-item-header")
    for name in movie_data:
        movie_name = name.find("a").get_text()
        movie_names.append(movie_name)
        
for component in movie_table_list:
    movie_year_data = component.find_all("h3", class_="lister-item-header")
    for year in movie_year_data:
        movie_year = year.find(class_="lister-item-year").get_text()
        movie_years.append(movie_year)

for discrip in movie_table_list:
    movie_discription = discrip.find_all("p", class_="")
    for desc in movie_discription:
        summary_text = desc.get_text().strip()
        summary.append(summary_text)

````

````python
random_df = pd.DataFrame({'Movie Names': movie_names,'Release Year': movie_years,'summary': summary})
````
